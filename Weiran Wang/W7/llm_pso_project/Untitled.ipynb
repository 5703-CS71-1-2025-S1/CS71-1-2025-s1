```python
import matplotlib.pyplot as plt

# 从图像中估算的 AUC 历史数据
auc_gpt = [0.7643, 0.7694, 0.7694, 0.7694, 0.7709, 0.7709, 0.7709, 0.7765, 0.7765, 0.7765]
auc_pso = [0.7481, 0.7602, 0.7630, 0.7714, 0.7714, 0.7714, 0.7714, 0.7714, 0.7714, 0.7714]

# 绘制对比图
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(auc_gpt)+1), auc_gpt, marker='o', label='GPT + PSO')
plt.plot(range(1, len(auc_pso)+1), auc_pso, marker='x', label='PSO Only')
plt.title('AUC Convergence Comparison')
plt.xlabel('Iteration')
plt.ylabel('Validation AUC')
plt.grid(True)
plt.legend()
plt.savefig("logs/auc_comparison_manual.png")
plt.show()
```
